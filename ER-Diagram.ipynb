```python
from graphviz import Digraph

# Skapa ett nytt Digraph-objekt
erd = Digraph(comment='ER-diagram för Bokhandel')

# Ange noder med etiketter (baserat på ditt tidigare DOT-exempel)
erd.node('F', 'Författare\nID (PK)\nFörnamn\nEfternamn\nFödelsedatum')
erd.node('B', 'Böcker\nISBN13 (PK)\nTitel\nSpråk\nPris\nUtgivningsdatum\nSidor\nFörlagID (FK)')
erd.node('BF', 'BokFörfattare\nBokISBN13 (PK, FK)\nFörfattareID (PK, FK)')
erd.node('FL', 'Förlag\nID (PK)\nNamn\nAdress\nStad\nPostnummer')
erd.node('BU', 'Butiker\nID (PK)\nButiksnamn\nAdress\nStad\nPostnummer')
erd.node('LS', 'LagerSaldo\nButikID (PK, FK)\nISBN13 (PK, FK)\nAntal')
erd.node('KU', 'Kunder\nID (PK)\nFörnamn\nEfternamn\nEmail\nTelefonnummer\nAdress')
erd.node('O', 'Ordrar\nOrderID (PK)\nKundID (FK)\nOrderDatum\nTotalBelopp')
erd.node('OR', 'Orderrader\nOrderID (PK, FK)\nISBN13 (PK, FK)\nAntal\nPrisAtSale')

# Lägg till relationer med pilar
erd.edge('B', 'FL', label='FörlagID')
erd.edge('BF', 'B', label='BokISBN13')
erd.edge('BF', 'F', label='FörfattareID')
erd.edge('LS', 'BU', label='ButikID')
erd.edge('LS', 'B', label='ISBN13')
erd.edge('O', 'KU', label='KundID')
erd.edge('OR', 'O', label='OrderID')
erd.edge('OR', 'B', label='ISBN13')

# Visa diagrammet
erd.render('er_diagram', view=True) 
erd  
```
